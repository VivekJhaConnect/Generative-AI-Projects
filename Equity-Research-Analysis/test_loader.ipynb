```python
from langchain.text_splitter import CharacterTextSplitter

splitter = CharacterTextSplitter(
    separator='\n',
    chunk_size=200,
    chunk_overlap=0
)

# splitter.split_text(data[0])
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

r_splitor = RecursiveCharacterTextSplitter(
    separators=['\n\n', '\n', ' '],
    chunk_size=200,
    chunk_overlap=0
)

chunks = r_splitor.split_text(data[0].page_content)
len(chunks)
```

```python
for chunk in chunks:
    print(len(chunk))
```
